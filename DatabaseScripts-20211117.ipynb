# [dbo].[PS_AUD_Audit_Trace]

```sql
/****** Object:  StoredProcedure [dbo].[PS_AUD_Audit_Trace]    Script Date: 2021/11/17 13:19:21 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO
CREATE PROCEDURE [dbo].[PS_AUD_Audit_Trace]	
AS
BEGIN
	-- SET NOCOUNT ON added to prevent extra result sets from
	-- interfering with SELECT statements.
	SET NOCOUNT ON;
	SELECT [aud].[aud_id],
	       [usr].[usr_name] + ' ' + [usr].[usr_surname]  as FullNames   
		  ,[aud].[aud_event_datetime]
		  ,[aud].[aud_event_name]
		  ,[aud].[aud_event_action]
	  FROM [dbo].[TM_AUD_Audit] aud INNER JOIN [dbo].[TM_USR_Users] usr ON aud.usr_id = usr.usr_id
END
GO
```

# [dbo].[PS_CATEGORIES]

```sql
/****** Object:  StoredProcedure [dbo].[PS_CATEGORIES]    Script Date: 2021/11/17 13:19:21 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO
CREATE PROCEDURE [dbo].[PS_CATEGORIES] 
  @cat_id varchar(50) = -1
AS
BEGIN
	SET NOCOUNT ON;
	
	IF (@cat_id = -1)
	BEGIN
	SELECT [cat_id]
          ,[cat_code]
          ,[cat_name]
          ,[cat_isactive]
      FROM [dbo].[TM_CAT_Categories]	
   END
   ELSE
   BEGIN
   SELECT  [cat_id]
          ,[cat_code]
          ,[cat_name]
          ,[cat_isactive]
      FROM [dbo].[TM_CAT_Categories]	
	  WHERE [cat_id] = @cat_id
   END
END
GO
```

# [dbo].[PS_LOGIN]

```sql
/****** Object:  StoredProcedure [dbo].[PS_LOGIN]    Script Date: 2021/11/17 13:19:21 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO

--EXEC PS_LOGIN 'shakings@hotmail.co.za','P@ssw0rd'
CREATE PROCEDURE [dbo].[PS_LOGIN]
	@email VARCHAR(50), @password VARCHAR(50)
	AS
BEGIN
 SET NOCOUNT ON;
 SELECT [usr_id]
       ,[usr_name]
       ,[usr_surname]
       ,[usr_email]
   FROM [dbo].[TM_USR_Users]
  WHERE [usr_email] =@email AND [usr_password] =@password  AND [usr_isactive]=1

END
GO
```

# [dbo].[PS_PRODUCTS]

```sql
/****** Object:  StoredProcedure [dbo].[PS_PRODUCTS]    Script Date: 2021/11/17 13:19:21 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO

CREATE PROCEDURE [dbo].[PS_PRODUCTS]
  @prd_id int = -1
AS
BEGIN
	SET NOCOUNT ON;
	IF (@prd_id = -1)
	BEGIN
	SELECT [prd].[prd_id]
          ,[prd].[prd_code]
          ,[prd].[prd_name]
          ,[prd].[prd_description]
          ,[prd].[prd_price]
          ,[prd].[prd_image]
		  ,[cat].[cat_id]
          ,[cat].[cat_name]
      FROM [dbo].[TM_PRD_Product] prd INNER JOIN [dbo].[TM_CAT_Categories] cat 
	    ON prd.cat_id = cat.cat_id
	 END
	 ELSE
	 BEGIN
	 SELECT [prd].[prd_id]
          ,[prd].[prd_code]
          ,[prd].[prd_name]
          ,[prd].[prd_description]
          ,[prd].[prd_price]
          ,[prd].[prd_image]
		  ,[cat].[cat_id]
          ,[cat].[cat_name]
      FROM [dbo].[TM_PRD_Product] prd INNER JOIN [dbo].[TM_CAT_Categories] cat 
	    ON prd.cat_id = cat.cat_id AND [prd].[prd_id] = @prd_id
	 END
END

--select * from [dbo].[TM_CAT_Categories]
GO
```

# [dbo].[PU_CATEGORIES]

```sql
/****** Object:  StoredProcedure [dbo].[PU_CATEGORIES]    Script Date: 2021/11/17 13:19:21 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO
CREATE PROCEDURE [dbo].[PU_CATEGORIES]
    @cat_id int,
    @cat_code varchar(50),
    @cat_name varchar(50),
	@cat_isactive int
AS
BEGIN
	SET NOCOUNT ON;
	UPDATE [dbo].[TM_CAT_Categories]
       SET [cat_code] = @cat_code
          ,[cat_name] = @cat_name
          ,[cat_isactive] = @cat_isactive
     WHERE [cat_id] = @cat_id

END
GO
```
